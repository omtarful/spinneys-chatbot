```python
import re
import requests
from bs4 import BeautifulSoup


def extract_urls_from_sitemap(sitemap_url, regex=None):
    # Send a GET request to the sitemap URL
    response = requests.get(sitemap_url)

    # Parse the XML content of the sitemap
    soup = BeautifulSoup(response.content, 'xml')

    # Find all <loc> tags in the sitemap
    loc_tags = soup.find_all('loc')

    # Extract the URLs from the <loc> tags
    urls = [loc.text for loc in loc_tags]

    # Filter the URLs based on the provided regex
    if regex:
        urls = [url for url in urls if re.search(regex, url)]

    return urls
```

```python
sitemap_url = "https://www.spinneyslebanon.com/media/sitemap-1-1.xml"
regex_pattern = r"^https://www.spinneyslebanon\.com/default/.*\.html$"
urls = extract_urls_from_sitemap(sitemap_url=sitemap_url, regex=regex_pattern)
urls
```

```python
sitemap_url = "https://www.spinneyslebanon.com/media/sitemap-1-2.xml"
regex_pattern = r"^https://www.spinneyslebanon\.com/default/.*\.html$"
urls_2 = extract_urls_from_sitemap(sitemap_url=sitemap_url, regex=regex_pattern)
urls_2
```

```python
urls = set(urls).union(urls_2)
```

```python
from langchain.document_loaders import UnstructuredURLLoader
loaders = UnstructuredURLLoader(urls=urls)

data = loaders.load()
```
